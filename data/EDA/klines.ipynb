```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd


# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]],subplot_titles=('OHLC', 'Volume'))

# include candlestick with rangeselector
fig.add_trace(go.Candlestick(x=df['open_time'],
                open=df['open_price'],
                high=df['high_price'],
                low=df['low_price'],
                close=df['close_price'],
                ),secondary_y=True)
               

# include a go.Bar trace for volumes
#fig.add_trace(go.Bar(x=df['open_time'], y=df['volume'],),
#               secondary_y=False)
fig2 = px.line(df.iloc[:-1], x='open_time', y='upper_BB',color_discrete_sequence=px.colors.qualitative.G10)
fig3 = px.line(df.iloc[:-1], x='open_time', y='lower_BB',color_discrete_sequence=px.colors.qualitative.Set1)
fig.add_trace(fig2.data[0],secondary_y=False)
fig.add_trace(fig3.data[0],secondary_y=False)

fig.layout.yaxis2.showgrid=False
# Do not show OHLC's rangeslider plot 
fig.update(layout_xaxis_rangeslider_visible=False)
fig.show()
```

```python
from plotly.colors import qualitative
# add colors for the 'value bar'
colors = []
for ind, val in enumerate(df['value']):
  if val == None or ind ==0:
    color = 'maroon'
  elif val >= 0:
    color = 'green'
    if val > df['value'][ind-1]:
      color = 'greenyellow'
  else:
    color = 'maroon'
    if val < df['value'][ind-1]:
      color='red'
  colors.append(color)

print(colors)
```

```python
fig = go.Figure(data=[go.Bar(
    x=df['open_time'],
    y=df['value'],
    marker_color=colors # marker color can be a single color value or an iterable
)])
fig.show()
```
